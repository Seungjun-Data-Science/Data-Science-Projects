```python
# Import relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in data
df = pd.read_csv("data_counsel_chat_with_dolly_answers.csv")
```

```python
# Calculate word length of each human counselor text and AI generated text
df['answerText_len'] = df['answerText'].apply(lambda x: len(x))
df['dollyAnswerText_len'] = df['dollyAnswerText'].apply(lambda x: len(x))
```

```python
# Calculate word count of each human counselor text and AI generated text
df['answerText_wc'] = df['answerText'].apply(lambda x: len(x.split()))
df['dollyAnswerText_wc'] = df['dollyAnswerText'].apply(lambda x: len(x.split()))
```

```python
# Standard Deviation of human counselor response text length
df['answerText_len'].std()
```

```python
# Standard Deviation of AI generated response text length
df['dollyAnswerText_len'].std()
```

```python
# Standard Deviation of human counselor response word count
df['answerText_wc'].std()
```

```python
# Standard Deviation of AI generated response word count
df['dollyAnswerText_wc'].std()
```

```python
# Number of cases where dolly response was longer than that of human counselor for each question asked
(df['answerText_len'] < df['dollyAnswerText_len']).sum()
```

```python
# Number of cases where dolly response contained more words than that of human counselor for each question asked
(df.answerText_wc < df.dollyAnswerText_wc).sum()
```

```python
# Distributions of response text length

fig, axes = plt.subplots(2)
sns.histplot(data = df, x ='answerText_len', ax=axes[0], color='orange')
sns.histplot(data = df, x = 'dollyAnswerText_len', ax=axes[1], color='orange')

axes[0].set_xlabel('Distribution of Human Counselor Response Text Length')
axes[1].set_xlabel('Distribution of Dolly Response Text Length')
plt.tight_layout()
```

```python
# Distributions of response text length

fig, axes = plt.subplots(2)
sns.histplot(data = df, x ='answerText_wc', ax=axes[0], color='green')
sns.histplot(data = df, x = 'dollyAnswerText_wc', ax=axes[1], color='green')

axes[0].set_xlabel('Distribution of Human Counselor Response Word Count')
axes[1].set_xlabel('Distribution of Dolly Response Word Count')
plt.tight_layout()
```
